```python
class number_check:
    def __init__(self, num):
        self.num = num
    
    def is_disarium(self):
        num_str = str(self.num)
        disarium_form = sum([int(i) ** (index + 1) for index, i in enumerate(num_str)])
        if disarium_form == self.num:
            return True
        else:
            return False
        
    def sum_of_digit_raised_to_power(self, n, p):
        '''
        Finds sum of the digits raised to a power.
        '''
        return sum([int(i) ** p for i in str(n)])
    
    def is_happy(self):
        n = self.num
        while True:
            n = self.sum_of_digit_raised_to_power(n, 2)
            if n < 10:
                break
        if n == 1:
            return True
        else:
            return False
        
    def is_harshad(self):
        sum_of_digits = self.sum_of_digit_raised_to_power(self.num, 1)
        if self.num % sum_of_digits == 0:
            return True
        else:
            return False
```

1. Write a Python program to check if the given number is a Disarium Number?

```python
n = number_check(518)
n.is_disarium()
```

2. Write a Python program to print all disarium numbers between 1 to 100?

```python
disarium_nums = []

for num in range(1, 101):
    if number_check(num).is_disarium():
        disarium_nums.append(num)
        
print(disarium_nums)
```

4. Write a Python program to print all happy numbers between 1 and 100?

```python
happy_nums = []

for num in range(1, 101):
    if number_check(num).is_happy():
        happy_nums.append(num)
        
print(happy_nums)
```
